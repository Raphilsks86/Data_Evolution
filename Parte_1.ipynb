```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Preparando os dados:

dados = pd.DataFrame({
    "Campanha" : ['Nike', 'Puma', 'Adidas', 'Reebook', 'New Balance'],
    "Cliques" : [320, 250, 300, 220, 280],
    "Conversões" : [40, 30, 35, 32, 34]
})
```

```python
# Configurando o gráfico:

plt.figure(figsize = (14, 4))
sns.scatterplot(data = dados, x = "Cliques", y = "Conversões", hue = "Campanha", palette = "cool")
```

```python
# Criando uma grade de visualização:

plt.figure(figsize = (14, 4))
sns.scatterplot(data = dados, x = "Cliques", y = "Conversões", hue = "Campanha", palette = "cool", markers = {"Cliques" : "x", "Conversões" : "y"})

plt.title("Relação entre cliques e conversões por campanha")
plt.xlabel("Cliques")
plt.ylabel("Conversões")
plt.grid(True)
plt.show()
```
